```python
# Libraries

import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd

df = pd.DataFrame(columns=['url','name','region','length','elevation','hp','currentrating','ratingcount','features','excerpt'])

```

HTML Request to get a Page

Update page number in line 1 as needed

Will need to loop across all pages as each page only has 30 trails

```python
for i in range(122):
    page = requests.get('https://www.wta.org/go-outside/hikes?b_start:int=' + str(i))
    contents = page.content
    #Save page.content to a file to avoid API hits
    soup = BeautifulSoup(contents, 'html.parser')
    alltrails = soup.findAll("div", {"class": "search-result-item"})
    for trail in alltrails:
        try:
            url = trail.div.div.a['href']
        except AttributeError:
            url = ''
        #print(url)
        try:
            name = trail.div.div.span.contents[0]
        except AttributeError:
            name = ''
        try:
            region = trail.div.div.h3.contents[0]
        except AttributeError:
            region = ''
        try:
            length = trail.div.find("div", {"class": "hike-length hike-stat"}).span.contents[0]
        except AttributeError:
            length = ''
        try:
            elevation = trail.div.find("div", {"class": "hike-gain hike-stat"}).span.contents[0]
        except AttributeError:
            elevation = ''
        try:
            hp = trail.div.find("div", {"class": "hike-highpoint hike-stat"}).span.contents[0]
        except AttributeError:
            hp = ''
        try:
            currentrating = trail.div.find("div", {"class": "current-rating"}).contents[0]
        except AttributeError:
            currentrating = ''
        try:
            ratingcount = trail.div.find("span", {"class": "rating-count"}).contents[0]
        except AttributeError:
            ratingcount = ''
        try:
            features_content = trail.div.findAll("div", {"class": "trip-features"})[0].findAll("img")
        except AttributeError:
            features_content = []
        features = []
        for x in features_content:
            features.append(x['title'])
        try:
            excerpt = trail.div.find("div", {"class": "listing-summary omega show-excerpt"}).contents[0]
        except AttributeError:
            excerpt = ''
        df = df.append({'url' : url , 'name' : name, 'region' : region,'length': length,'elevation': elevation,'hp':hp,'currentrating':currentrating,'ratingcount' :ratingcount,'features':features,'excerpt':excerpt} , ignore_index=True)
        
```

Initialize df_trip if running in batch mode

```python
df_trip = pd.DataFrame(columns=['trip_url','hike_url','hike_author','hike_date','hike_text','type_of_hike','trail_conditions','road','bugs','snow'])
```

```python
from IPython.display import clear_output

def update_progress(iteration, progress):
    bar_length = 20
    if isinstance(progress, int):
        progress = float(progress)
    if not isinstance(progress, float):
        progress = 0
    if progress < 0:
        progress = 0
    if progress >= 1:
        progress = 1
    block = int(round(bar_length * progress))
    clear_output(wait = True)
    text = "Progress: [{0}] {1:.1f}%".format( "#" * block + "-" * (bar_length - block), progress * 100, iteration)
    print(text)
    print("Iteration: "+ str(iteration))
    
```

```python
start = 150127
batchsize = len(trip_urls['URLs'])
for i, trip in enumerate(trip_urls['URLs'], start=start):
    update_progress(i,(i)/batchsize)
    #print(i)
    page = requests.get(trip)
    contents = page.content
    soup = BeautifulSoup(contents, 'html.parser')
    trip_url = trip
    hike_url = soup.find("h1", {"class": "documentFirstHeading"})
    try:
        hike_url = hike_url.a['href']
    except (AttributeError, TypeError):
        hike_url = ''

    hike_author = soup.find("div", {"class": "CreatorInfo"})
    try:
        hike_author = hike_author.span.a['href']
    except (AttributeError, TypeError):
        hike_author = ''

    hike_date = soup.find("span", {"class": "elapsed-time"})
    try:
        hike_date = hike_date['datetime']
    except (AttributeError, TypeError):
        hike_date = ''

    hike_text = soup.find("div", {"id": "tripreport-body-text"})
    try:
        hike_text = hike_text.p.contents[0]
    except (AttributeError, TypeError, IndexError):
        hike_text = ''

    trip_conditions = soup.findAll("div", {"class": "trip-condition"})
    type_of_hike = ''
    trail_conditions = ''
    road = ''
    bugs = ''
    snow = ''
    if len(trip_conditions) != 0:

        
        for condition in trip_conditions:
            if condition.h4.contents[0] == 'Type of Hike':
                type_of_hike = condition.span.contents[0]
                #print(type_of_hike)
            if condition.h4.contents[0] == 'Trail Conditions':
                trail_conditions = condition.span.contents[0]
                #print(trail_conditions)
            if condition.h4.contents[0] == 'Road':
                road = condition.span.contents[0]
                #print(road)
            if condition.h4.contents[0] == 'Bugs':
                bugs = condition.span.contents[0]
                #print(bugs)
            if condition.h4.contents[0] == 'Snow':
                snow = condition.span.contents[0]
                #print(snow)
    
    df_trip.loc[0, ['trip_url']] = trip_url
    df_trip.loc[0, ['hike_url']] = hike_url
    df_trip.loc[0, ['hike_author']] = hike_author
    df_trip.loc[0, ['hike_date']] = hike_date
    df_trip.loc[0, ['hike_text']] = str(hike_text)
    df_trip.loc[0, ['type_of_hike']] = type_of_hike
    df_trip.loc[0, ['trail_conditions']] = trail_conditions
    df_trip.loc[0, ['road']] = road
    df_trip.loc[0, ['bugs']]= bugs
    df_trip.loc[0, ['snow']] = snow
    
    with open(r'tripreports.csv', 'a') as f:
        df_trip.to_csv(f, header=False, index=False, index_label=False)
    #with open(r'tripreports.csv', 'a') as f:
    #writer = csv.writer(f)
    #for x in trip_report_urls: 
    #    writer.writerow([x])

update_progress(start+batchsize-1,1)
```
